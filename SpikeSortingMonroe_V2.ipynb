```python
for key in (ss.SpikeSorting() & key).fetch('KEY'):
    key['waveform_params_name'] = waveform_params_name
    we = (ss.Waveforms & key).load_waveforms(key)
    waveforms_w_ind = [we.get_waveforms(unit,with_index=True) for unit in we.sorting.get_unit_ids()]
    spike_waveforms = [np.swapaxes(waveforms[0], 0, 2) for waveforms in waveforms_w_ind]
    spike_inds = [np.array(list(zip(*waveforms[1])),dtype=np.int64)[0] for waveforms in waveforms_w_ind]
    spike_times = [we.sorting.get_unit_spike_train(unit_id=unit) for unit in we.sorting.get_unit_ids()]
    selected_spike_times = [spike_times[unit][inds] for unit, inds in enumerate(spike_inds)]
    n_clusters = len(we.sorting.get_unit_ids())
    n_channels = len(we.recording.get_channel_ids())
    spike_waveform_arr = make_waveform_arr(spike_waveforms, n_clusters, n_channels, 28)
    spike_times_arr = make_spike_times_arr(selected_spike_times, n_clusters)
    spike_window_min = np.nanmin(spike_waveform_arr[:,:,14-3:14+3,:], axis=2)
    rec_timestamps = we.recording.get_times()
    samp_freq = we.recording.get_sampling_frequency()
    spike_amplitude_dict[key['sort_group_id']] = {'spike_times' : spike_times_arr,
                                                  'spike_amplitude' : spike_window_min,
                                                  'samp_freq' : samp_freq,
                                                  'timestamps' : rec_timestamps}
```
